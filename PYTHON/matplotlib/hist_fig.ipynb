直方图

```python
import numpy as np
import matplotlib.pyplot as plt
N=1000
np.random.seed(100)
x=np.random.randn(N)*2+15

plt.figure(figsize=(18,12))
plt.hist(x,bins=25,range=(12,18),color='royalblue',orientation='horizontal',edgecolor='k',label='$mu=15,\sigma=2$')
plt.legend(prop={'family':'smihei','size':14})
```
