线图

```python
import numpy as np
import matplotlib.pyplot as plt

N = 50
x=np.linspace(0.,10.,N)
y=np.sin(x)**2 + np.cos(x)

rows = 2
cols = 2

grid=plt.GridSpec(rows,cols,wspace=.25,hspace=.25)
# 几种线条样式,似乎只有这四种
linestyles=['-','--','-.',':']
color = ['#00429d', '#627c94', '#f4777f', '#93003a']

# 每张图的大小，放在循环外边似乎更好一点
plt.figure(figsize=(17,9))
for i in range(len(linestyles)):
    # plt.subplot(2,2,i+1)
    plt.subplot(grid[i])
    # x是标记点的形状，ms size
    plt.plot(x,y,'x',linestyle=linestyles[i],label=r'$y=sin^2(x)+cos(x)$',ms=7,markevery=5,color=color[i])
    plt.axis('equal')
    plt.legend()
    plt.annotate('linestyle:'+str(linestyles[i]),xy=(0.5,-2.5),va='center',ha='left')

```

```python
import numpy as np
import matplotlib.pyplot as plt

N=25

x=np.linspace(0.,10.,N)
y=np.sin(x)**2 + np.cos(x)

np.random.seed(100)
noise = np.random.random(N)*.7+.4

plt.figure(figsize=(14,9))
plt.axis('equal')
plt.plot(x,y,ls='-',label=r'$y=sin^2(x)+cos(x)$',color='darkgreen')
plt.fill_between(x,y+noise,y-noise,color='darkgreen',alpha=.5)

plt.hlines(0,xmin=0,xmax=10,ls='--',color='royalblue',label='hlines')
plt.vlines(2,ymin=-3,ymax=3,ls='--',color='orange',label='vlines')

# bbox_ num1 左右移动 num2上下移动，值越大，向右上移动越多
# loc默认是1，用来换锚点位置的？
plt.legend(loc = 3,bbox_to_anchor=(0,0),prop={'family':'smihei','size':14})
```

一张图绘制多个图像，调用多次plt就可以

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

或者可以改变线条和颜色

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(x,x+0,linestyle='solid', color='blue')
plt.plot(x,x+1,linestyle='dashed', color='g')
plt.plot(x,x+2,linestyle='dashdot', color='0.75')
plt.plot(x,x+3,linestyle='dotted', color='#FFDD44')

plt.plot(x,x+4,linestyle='-')
plt.plot(x,x+5,linestyle='--')
plt.plot(x,x+6,linestyle='-.')
plt.plot(x,x+7,linestyle=':')
```
